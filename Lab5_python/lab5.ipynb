При решении научных задач с помощью численных методов зачастую приходится иметь дело с большими массивами данных в виде массивов и матриц. Постройте поточечный график функции. Сгенерируйте два массива для этого. Один со значениями по возрастанию (для X), второй со случайными значениями (для Y).



```python
import math
import matplotlib.pyplot as mplt
%matplotlib inline

x = np.arange(10)
y = np.random.random(10)

print(mplt.plot(x, y, 'o'))
```

Выберите тригонометрическую функцию $T(x)$, выберите полином $P(x)$. Постройте график функции $T(x)P(x)$.

```python
x = np.linspace(-3*np.pi, 3*np.pi, 200)
T = np.cos(x) + 1
P = 2*x + 3*x*x
fig, ax = mplt.subplots()
ax.plot(x, T)
ax.plot(x, P)
mplt.show()
```

К графику добавьте: обозначение оси абсцисс, обозначение оси ординат, название графика, вставка легенды (текста в label).

```python
x = np.linspace(-3*np.pi, 3*np.pi, 200)
T = np.cos(x) + 1
P = 2*x + 3*x*x
fig, ax = mplt.subplots()
ax.plot(x, T, label = "sin(x)")
ax.plot(x, P, label = "2*x + 4*x*x")
mplt.title("T(x)P(x)") 
mplt.xlabel("Ox")
mplt.ylabel("Oy") 
ax.legend()
```

Измените цвет графика на Ваш любимый. 

```python
x = np.linspace(-3*np.pi, 3*np.pi, 200)
T = np.cos(x) + 1
P = 2*x + 3*x*x
fig, ax = mplt.subplots()
ax.plot(x, T, label = "sin(x)", color="blue")
ax.plot(x, P, label = "2*x + 4*x*x", color="red")
mplt.title("T(x)P(x)") 
mplt.xlabel("Ox")
mplt.ylabel("Oy") 
ax.legend()
```

Постройте несколько квивых. Первая - цвет красный, сплошная, вторая - цвет синий, штрих-пунктир.

```python
x = np.linspace(-3*np.pi, 3*np.pi, 200)
T = np.sin(x)
P = np.cos(x)
fig, ax = mplt.subplots()
ax.plot(x, T, color = "red", linestyle = "-")
ax.plot(x, P, color = "blue", linestyle = "--")
```

Постройте теже графики, только для первого добавьте маркеры из красных треугольников, для второго синие круглые маркеры.

```python
x = np.linspace(-3*np.pi, 3*np.pi, 200)
T = np.sin(x)
P = np.cos(x)
fig, ax = mplt.subplots()
ax.plot(x, T, color = "red", marker = 6)
ax.plot(x, P, color = "blue", marker = "8")
fig.savefig('graph')
```

Прочитайте картинку используя opencv

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('img1.jpg')
mplt.imshow(img)
```

Сделайте ее черно-белой

```python
ImgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(ImgGray)
```

Измените размер на 300x300

```python
imag = cv2.resize((img), (300, 300))
plt.imshow(imag)
```

Разбейте ее на 2 картинки. Сохраните. 

```python
(h, w) = imag.shape[:2]
left = imag[:h, :w//2]
right = imag[:h, w//2:]
cv2.imwrite('imgLeft.png', left)
cv2.imwrite('imgRight.png', right)
```
